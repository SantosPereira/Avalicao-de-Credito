# Predizendo aprovação de crédito

```python
import sklearn
import pandas as pd
```

```python
df = pd.read_csv('credit_customers.csv')
```

```python
df
```

```python
df.describe()
```

```python
df.dropna().describe()
```

```python
df.dropna(inplace=True)
```

```python
df = pd.get_dummies(df)
```

```python
df
```

```python
# Separando treino e teste
train = df.sample(frac=0.8, random_state=200)
test = df[~df.index.isin(train.index)]

x_train, y_train = train.drop(columns=['class_good', 'class_bad']), train['class_good']
x_test, y_test = test.drop(columns=['class_good', 'class_bad']), test['class_good']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
predito = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test.astype(int), predito.astype(int))
```
